**Importamos librerias**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Leemos el .csv**

```python
df= pd.read_csv ('Dallas.csv')
df
```

```python
df. info()
```

**CREAMOS LOS FILTROS**

```python
#Los registros de 5 "host name" diferentes 
punto_a= df[df["host_name"].isin(["Sasha", "Robert", "Judy", "Joan", "Rc"])]
punto_a
```

```python
#Los registros de los host que se hallan unido a Airbnb "antes del año 2019"
punto_b= df[df["host_since"] < "2019-01-01"]
punto_b
```

```python
#Los registros de los host que responden "a mas tardar en 1 dia"
punto_c= df[df["host_response_time"].isin(["within an hour", "within a few hours", "within a day"])]
punto_c
```

```python
#Los registros que tengan "al menos 5 acomadantes"
punto_d= df[(df["accommodates"] <= 5)]
punto_d
```

```python
#Los registros de los tipos de cuarto: "Entire home" y "Private room"
punto_e= df[(df["property_type"] == "Entire home") | (df["room_type"] == "Private room")]
punto_e
```

```python
#Los registros que "a lo mucho cuenten con 4 camas"
punto_f= df[(df["beds"] <= 4)]
punto_f
```

```python
#los registros que "tengan 1 baño o que sea superhost" pero con la condicion de que "todos sean instant_bookable"
punto_g= df [(df["bathrooms"] == "1") | (df["host_is_superhost"] == "t") & (df["instant_bookable"] == "t")]
punto_g
```

```python
#Los registros cuyo "review_scores_cleanliness sea menor de 3 y review_scores_rating sea mayor de 3"
punto_h= df[(df["review_scores_cleanliness"] < 3) & (df["review_scores_rating"] > 3)] 
punto_h
```
